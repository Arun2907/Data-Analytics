```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# uploading the dataset

df=pd.read_excel('Data_Train.xlsx')
df
```

```python
#statistical view

df.describe()
```

```python
df.sample(15)
```

```python
#seperate date from date of journey

df['Date']=pd.to_datetime(df.Date_of_Journey,format='%d/%m/%Y')
df.head()
```

```python
#seperate month from date of journey

df['month']=pd.to_datetime(df.Date_of_Journey,format='%d/%m/%Y').dt.month
df.head()
```

```python
#droping the date of journey column

df.drop(['Date_of_Journey'],axis=1,inplace=True)
df
```

```python
#seperate day from date 

df['day']=pd.to_datetime(df.Date,format='%d/%m/%Y').dt.day
df.head()
```

```python
#converting duration into minutes

l=list(df['Duration'])
h=0
m=0
d=[]
for i in l:
    s=str(i).split(' ')
    if len(s)==2:
        if 'h' in s[0]:
            h=s[0].split('h')
        if 'm' in s[1]:
            m=s[1].split('m')
            
    if len(s)==1:
        if 'h' in s[0]:
            h=s[0].split('h')
        if 'm' in s[0]:
            m=s[0].split('m')
            
    d.append((int(h[0])*60+int(m[0])))
    
    h=[0]
    m=[0]
    
d
        
```

```python
df['minutes']=d
df[['Duration','minutes']].sample(20)
```

```python
#convert depature time into time format

df['dep_time']=pd.to_datetime(df.Dep_Time,format='%H:%M').dt.time
df
```

```python
df.info()
```

```python
#airlines wise maximum based on price

airlines=df.groupby('Airline')
df_mp=pd.DataFrame(airlines['Price'].max())
df_mp
df1=df[['Airline','Source','Destination','Price','Route']]
df2=pd.merge(df_mp,df1,left_on=['Airline','Price'],right_on=['Airline','Price'])
df2
df2.drop_duplicates()
```

```python
#airlines wise based on duration of maximum minutes

airlines=df.groupby('Airline')
df_mp=pd.DataFrame(airlines['minutes'].max())
df_mp
df1=df[['Airline','Source','Destination','minutes','Route']]
df2=pd.merge(df_mp,df1,left_on=['Airline','minutes'],right_on=['Airline','minutes'])
df2
df2.drop_duplicates().reset_index()
```

```python
#making heatmap by airline and mean price

df1=df[['Airline','month','Price']]
df2=df1.groupby(['Airline','month'])
df3=pd.DataFrame(df2['Airline','Price'].mean()).reset_index().round()
df4=df3.pivot('Airline','month','Price')
df4
sns.heatmap(df4,cmap='BuGn',annot=True,fmt='.1f')
```
