```python
sns.distplot(df['percentage'],color='green')
plt.title('Comparison of percentage scored by all the student')
```

```python
def grades(p):
    
    if(p>=90):
        return 'o'
    if(p>=80):
        return 'A'
    if(p>=70):
        return 'B'
    if(p>=60):
        return 'C'
    if(p>=50):
        return 'D'
    if(p>=40):
        return 'E'
    else:
        return 'F'
    
df['status']=df.apply(lambda x:grades(x['percentage']),axis=1)

    
```
