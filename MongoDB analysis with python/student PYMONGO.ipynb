```python
#importing mongodb

from pymongo import MongoClient
```

```python
client = MongoClient('mongodb://localhost:27017/')
```

```python
#listing databases

client.list_database_names()
```

```python
#choosing the database

db = client ["student"]
```

```python
#listing collections

db.list_collection_names()
```

```python
# count of records

db.test.count_documents({})
```

```python
#showing records in database

for i in db.test.find():
    print(i)
```

```python
#overall percentage

r=db.test.aggregate([{"$group":{"_id":"null","overall":{"$avg":"$percentage"}}}])

for i in r:
    print(i)
```

```python
#genderwise average

r=["male","female"]
for i in r:
    p=db.test.aggregate([{"$match":{"gender":i}},{"$group":{"_id":i,"overall":{"$avg":"$percentage"}}}])
    for j in p:
        print(j)
```

```python
#race/ethnicity overall

p=db.test.distinct("race/ethnicity")
for i in p:
    b=db.test.aggregate([{"$match":{"race/ethnicity":i}},{"$group":{"_id":i,"overall":{"$avg":"$percentage"}}}])
    for j in b:
        print(j)
```
